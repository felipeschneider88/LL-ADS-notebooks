## Prueba de código más complejo

Vamos a intentar obtener una imagen al azar usando una API de perros

```python
#Agrego todas las librerias necesasrias
from PIL import Image
from io import BytesIO
import requests
import json

urlPerro = 'https://dog.ceo/api/breeds/image/random'

imgUri = json.loads(requests.get(urlPerro).content)["message"]

arreByte = requests.get(imgUri)

img = Image.open(BytesIO(arreByte.content))

display(img)

```
